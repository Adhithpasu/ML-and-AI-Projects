```python
covid_df["total_cases"] = covid_df["new_cases"].cumsum()
```

```python
locations_df.loc[locations_df.location == "Italy"]
```

```python
covid_df["location"] = "Italy"
```

```python
merged_df = covid_df.merge(locations_df, on="location")
```

```python
merged_df
```

```python
merged_df["cases_per_million"] = merged_df.total_cases * 1e6 / merged_df.population
```

```python
merged_df["deaths_per_million"] = merged_df.total_deaths * 1e6 / merged_df.population
```

```python
merged_df["tests_per_million"] = merged_df.total_deaths * 1e6 / merged_df.population
```

```python
merged_df
```

```python
result_df = merged_df[["date", "new_cases", "new_deaths", "new_tests", "total_cases", "total_deaths", "total_tests", "population", "cases_per_million", "deaths_per_million", "tests_per_million"]]
```
