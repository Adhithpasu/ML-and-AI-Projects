Histograms:

```python
flower_df = sns.load_dataset("iris")
flower_df.sepal_width
```

```python
flower_df.describe()
```

```python
plt.title("Distribution of Sepal Width")
plt.hist(flower_df.sepal_width); # This shows the distribution of data and for now it is split into 10 intervals but we can change it (Terminology: An interval also means a bin)
```

```python
plt.hist(flower_df.sepal_width, bins=5);
```

```python
import numpy as np
np.arange(2, 4.75, 0.25) # This is the np array that we pass into hist to get the equal length intervals
```

```python
plt.hist(flower_df.sepal_width, bins=np.arange(2, 4.75, 0.25)); # Seperating the bins into equal sizes
```

```python
plt.hist(flower_df.sepal_width, bins=[1, 1.5, 2, 2.5, 3.9, 4, 6]); # You can also have unequal bin sizes
```

```python
setosa_df = flower_df[flower_df.species == "setosa"]
versicolor_df = flower_df[flower_df.species == "versicolor"]
virginica_df = flower_df[flower_df.species == "virginica"]
```

```python
plt.hist(setosa_df.sepal_width, bins=np.arange(2, 4.75, 0.25), alpha=0.5)
plt.hist(versicolor_df.sepal_width, bins=np.arange(2, 4.75, 0.25), alpha=0.5)
plt.legend(["Setosa", "Versicolor"]);
```

```python
plt.title("Distribution of different species sepal width")
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], bins=np.arange(2, 4.75, 0.25), stacked=True) # By setting stacked to True you can see all the species without worrying about overlapping since they are added on top of each other
plt.legend(["Setosa", "Versicolor"]);
```

```python
tip_df = sns.load_dataset("tips")
tip_df
```

```python
AverageBillByDay_df = tip_df.groupby("day")[["total_bill"]].mean() # Warning: if you do not put double square brackets on the "total_bill" then it will be considered a Series object not a dataframe object which we won't be able to display on a bar chart
```

```python
AverageBillByDay_df
```

```python
plt.bar(AverageBillByDay_df.index, AverageBillByDay_df.total_bill)
```

```python
sns.barplot(data = tip_df, x="day", y="total_bill") # The lines you find at the top of each bar is where most of the data lies within that range (Also called Error bars also known as confidence intervals, it is generally around the mean of all the values)
```

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tip_df)
```

```python
sns.barplot(y="day", x="total_bill", data=tip_df, hue="smoker") # You can flip the axises
```

```python
flight_df = sns.load_dataset("flights")
flight_df
```

```python
flight_df = sns.load_dataset("flights").pivot(columns="year", index="month", values="passengers") # The values keyword argument is not required but you need to specifiy the index and columns keyword arguments if you want this matrix looking layout
flight_df
```

```python
plt.title("Number of Passengers in Thousands")
sns.heatmap(flight_df);
```

```python
plt.title("Number of Passengers in Thousands")
sns.heatmap(flight_df, fmt="d", annot=True); # The annot will put in the numbers when set to true and the fmt is used to format the numbers (if you want decmals or integers use "d" or you can use scientfic notation then use "0.2f")
```

```python
plt.title("Number of Passenger in Thousands")
sns.heatmap(flight_df, fmt="d", annot=True, cmap="Blues") # The cmap keyword allows you to use inbuilt formatting for the heatmaps
```

```python
fig, axes = plt.subplots(2,3, figsize=(17, 9)) # The figsize keyword lets you ajust the size of all the graphs
axes[0,0].plot(years, oranges, "o--r", color="orange")
axes[0,0].plot(years, apples, "x-r")
axes[0,0].set_xlabel("Year")
axes[0,0].set_ylabel("Yields")
axes[0,0].set_title("Yields of Oranges") # The object(<Axes>) uses different named methods for some methods

axes[0,1].set_title("Sepal Length vs Sepal Width")
sns.scatterplot(x=flower_df.sepal_length, y=flower_df.sepal_width, hue = flower_df.species, s=100, ax=axes[0,1])

setosa_df = flower_df[flower_df.species == "setosa"]
versicolor_df = flower_df[flower_df.species == "versicolor"]
virginica_df = flower_df[flower_df.species == "virginica"]
axes[0,2].set_title("Sepal Width Distribution")
axes[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], bins=np.arange(2, 5, 0.25), stacked = True)
axes[0,2].legend(["Setosa", "Versicolor", "Virginica"])

axes[1,0].set_title("Restaurant Bills")
sns.barplot(x="day", y="total_bill", hue = "sex", data=tip_df, ax=axes[1,0])

axes[1,1].set_title("Flight traffic")
sns.heatmap(data=flight_df, cmap="Blues", ax=axes[1,1])

axes[1,2].set_title("Chart")
axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].set_xticks([]) # This removes the x axis by setting it into an empty array
axes[1,2].set_yticks([]) # This removes the y axis by setting it into an empty array

plt.tight_layout(pad=2) # This lets you adjust the padding between the numbers of each graph
```
