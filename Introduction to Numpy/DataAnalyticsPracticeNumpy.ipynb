```python
import numpy as np
a1 = list(range(1000000))
a2 = list(range(1000000, 2000000))
n1 = np.array(a1)
n2 = np.array(a2)
```

```python
%time
res = 0
for e, e2 in zip(a1, a2):
  res += (e * e2)
res
```

```python
%time
np.dot(n1, n2)
```
